```python
def return_last_90d_price(symbol):
  """
   - return the last 90 days of the stock prices of symbol as a list of json strings
  """
  vantage_api_key = userdata.get('vantage_api_key')
  url = f'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol={symbol}&apikey={vantage_api_key}'
  r = requests.get(url)
  data = r.json()
  results = []   # empyt list for now to hold the 90 days of stock info (open, high, low, close, volume)
  for d in data["Time Series (Daily)"]:  # here d is a date: "YYYY-MM-DD"
    stock_info = data["Time Series (Daily)"][d]
    stock_info["date"] = d
    results.append(stock_info)
    # results.append(data["Time Series (Daily)"][d])
    # an example of data["Time Series (Daily)"][d] is
    # {'1. open': '117.3500', '2. high': '119.6600', '3. low': '117.2500', '4. close': '117.8700', '5. volume': '286038878'}
  return results
```

```python
price_list = return_last_90d_price("NVDA")
price_list
```

```python
import snowflake.connector

# Snowflake connection
conn = snowflake.connector.connect(
    user= user_id,
    password=password,
    account=account,
    warehouse='my_warehouse',
    database='my_database',
    schema='my_database.raw_data'
)
```

```python
create_table_query = """
CREATE OR REPLACE TABLE raw_data.stock_prices (
    open FLOAT,
    high FLOAT,
    low FLOAT,
    close FLOAT,
    volume BIGINT,
    date DATE PRIMARY KEY
)
"""
```

```python
cur = conn.cursor()
cur.execute(create_table_query)
cur.close()
```

```python
def insert_stock_data_to_snowflake(data):
    cur = conn.cursor()

    # Prepare the SQL INSERT query without the 'symbol' column
    insert_query = """
    INSERT INTO raw_data.stock_prices (open, high, low, close, volume, date)
    VALUES (%(open)s, %(high)s, %(low)s, %(close)s, %(volume)s, %(date)s)
    """

    # Loop through the data, map the correct keys and insert into Snowflake
    for record in data:
        record_to_insert = {
            'open': record['1. open'],
            'high': record['2. high'],
            'low': record['3. low'],
            'close': record['4. close'],
            'volume': record['5. volume'],
            'date': record['date']  # Make sure 'date' is included
        }
        cur.execute(insert_query, record_to_insert)

    cur.close()

# Call the insert function
insert_stock_data_to_snowflake(price_list)
```

```python
def load_stock_data_with_transaction(data):
    try:
        # Start a transaction
        cur = conn.cursor()
        cur.execute("BEGIN;")  # Start the transaction

        # Create the table if it doesn't exist
        create_table_query = """
        CREATE OR REPLACE TABLE raw_data.stock_prices (
            date DATE PRIMARY KEY,
            open FLOAT,
            high FLOAT,
            low FLOAT,
            close FLOAT,
            volume BIGINT
        );
        """
        cur.execute(create_table_query)
        print("Table created successfully!")

        # Prepare the SQL INSERT query
        insert_query = """
        INSERT INTO raw_data.stock_prices (date, open, high, low, close, volume)
        VALUES (%(date)s, %(open)s, %(high)s, %(low)s, %(close)s, %(volume)s)
        """

        # Insert records into the table
        for record in data:
            record_to_insert = {
                'open': record['1. open'],
                'high': record['2. high'],
                'low': record['3. low'],
                'close': record['4. close'],
                'volume': record['5. volume'],
                'date': record['date']  # Ensure date is present
            }
            cur.execute(insert_query, record_to_insert)

        cur.execute("COMMIT;")  # Commit the transaction
        print("Data inserted successfully!")

    except Exception as e:
        # If there's an error, roll back the transaction
        cur.execute("ROLLBACK;")
        print(f"An error occurred: {e}")

    finally:
        cur.close()
```

```python
load_stock_data_with_transaction(price_list)
```

```python
def check_idempotency(data):
    # Get the count of records before running the insertion
    cur = conn.cursor()
    cur.execute("SELECT COUNT(*) FROM raw_data.stock_prices;")
    count_before = cur.fetchone()[0]
    print(f"Records before insertion: {count_before}")

    # Insert the data
    load_stock_data_with_transaction(data)

    # Get the count of records after running the insertion
    cur.execute("SELECT COUNT(*) FROM raw_data.stock_prices;")
    count_after = cur.fetchone()[0]
    print(f"Records after insertion: {count_after}")

    if count_before == count_after:
        print("Idempotency test passed: No duplicate records were inserted.")
    else:
        print(f"Idempotency test failed: {count_after - count_before} new records were inserted.")

    cur.close()

# Call the function twice to test idempotency
price_list = return_last_90d_price('NVDA')
check_idempotency(price_list)  # First run
check_idempotency(price_list)  # Second run
```
